###Sentiment Analysis using Transformer Models

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a hugging course my whole life.",
        "I hate this so much"
    ]

)
```

###Text Generation Model (like chatgpt)

```python
generator=pipeline ("text-generation")
generator("once upon a time in nuwakot....")
```

```python
qa=pipeline("question-answering")
qa(
   question="where do i live?" ,
   context="once upon a time in LA..",
)
```

###Language translator

```python
translator=pipeline("translation_en_to_de")
translator("I am studying LLM")
```

###NER Models

```python
ner=pipeline("ner",model="dbmdz/bert-large-cased-finetuned-conll03-english")
("I am Aayushka and I am from Kathmandu")
```

```python
ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am Aayushka Dahal and I am from Kathmandu,Nepal.")
```
